```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
    
# %matplotlib notebook
```

```python
points = np.random.normal(size=[1000, 2], loc=[2, 2], scale=1)  # Generating random numbers with (2, 2) as mean and sd=1
plt.scatter(points[:, 0], points[:, 1], alpha=0.3)
```

```python
hist, xedges, yedges = np.histogram2d(points[:, 0], points[:, 1], range = [[-1, 5], [-1, 5]])

plt.imshow(hist, interpolation='nearest', origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])

# The histogram
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
xs, ys = np.meshgrid(xedges[:-1], yedges[:-1])
xs, ys = xs.ravel(), ys.ravel()
zs = 0

dx = dy = 0.5 * np.ones_like(zs)
dz = hist.ravel()

ax.bar3d(xs, ys, zs, dx, dy, dz)
```

```python
print(f"Number of nodes with negative x {color.BOLD}or{color.END} y:", np.logical_or(points[:,0] < 0, points[:,1] < 0).sum())
print(f"Number of nodes with negative x {color.BOLD}and{color.END} y:", np.logical_and(points[:,0] < 0, points[:,1] < 0).sum())
```

```python
avg = np.mean(points, axis=0)
print(f"average:", avg)
print("Euclidean of the real and expected average:", np.linalg.norm(avg - [2, 2]))
```

```python
norms = np.linalg.norm(points - [2, 2], axis=1)
print("The nearest node to the expected average [2, 2] is ", points[norms ==  norms.min()][0])
```
