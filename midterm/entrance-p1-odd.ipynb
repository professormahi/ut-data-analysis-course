```python
import numpy as np
np.random.seed(0)
a=np.random.randn(10,3)
```

```python
a
```

```python
distance_mat = np.linalg.norm(a[:, None, :] - a[None, :, :], axis=-1)
```

```python
distance_mat
```

```python
distance_mat[distance_mat == 0] = np.inf
distance_mat
```

```python
nn = distance_mat.argmin(axis=0)
print(nn)
```
