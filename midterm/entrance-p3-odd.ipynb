```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('income.csv')
```

```python
df
```

```python
plt.figure(figsize=(6, 6))

male_incomes = df[df["gender"] == "Male"].groupby("income").count()["age"]
female_incomes = df[df["gender"] == "Female"].groupby("income").count()["age"]

plt.bar(male_incomes.index, male_incomes, label="Males")
plt.bar(female_incomes.index, female_incomes, label="Females")

plt.xlabel("income")
plt.ylabel("count")
plt.legend()
```

```python
mar_stat = df.groupby("marital-status").sum()["age"]
plt.bar(mar_stat.index, mar_stat)
plt.xticks(rotation=60);
```

```python
for stat in df["marital-status"].unique():
    plt.figure()
    plt.hist(df[df["marital-status"] == stat]["age"])
    plt.title(f"marital-status: {stat}")
    plt.xlabel("age")
    plt.ylabel("freq")
    plt.xlim(0, 90)
```
