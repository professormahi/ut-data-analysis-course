## Load Electricity data

```python
electricity = pd.read_csv('electricity_use_per_person.csv', index_col='country')
electricity.columns = [int(col) for col in electricity.columns]
display(electricity)

print("indexes:", electricity.index)
print("cols:", electricity.columns)
```

## Load oil data

```python
oil = pd.read_csv('oil_consumption_per_cap.csv', index_col='country')
oil.columns = [int(col) for col in oil.columns]
display(oil)

print("indexes:", oil.index)
print("cols:", oil.columns)
```

## Columns are different

```python
electricity = electricity.loc[oil.index, 1971:2014]
oil = oil.loc[:, 1971:2014]
display(electricity, oil)
```

## Some values are NaN

```python
electricity = electricity.fillna(method='bfill', axis=1)
oil = oil.fillna(method='bfill', axis=1)
display(electricity, oil)
```

## Visualize Data

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1, 2, 1)
plt.imshow(oil)
plt.subplot(1, 2, 2)
plt.imshow(electricity)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2, 1, 1)
plt.plot(oil.loc['Iran', :])
plt.subplot(2, 1, 2)
plt.plot(electricity.loc['Iran', :])
```

## Sort by usages mean

```python
electricity.loc[electricity.mean(axis=1).sort_values().index,]
```

```python
electricity['mean_usage'] = electricity.mean(axis=1)
electricity.sort_values(by='mean_usage')
```

## Iran vs World!

```python
if 'mean_usage' in electricity.columns:
    electricity = electricity.drop(['mean_usage'], axis=1)
plt.plot(electricity.mean(axis=0), label='World')
plt.plot(electricity.loc['Iran', :], label='Iran')
plt.xlabel('Year')
plt.ylabel('Electricity consumption per capita (KWh/year)')
plt.title('Electricity consumption, Iran vs the World')
plt.legend()
# plt.grid()
```

```python
if 'mean_usage' in oil.columns:
    electricity = oil.drop(['mean_usage'], axis=1)
plt.plot(oil.mean(axis=0), label='World')
plt.plot(oil.loc['Iran', :], label='Iran')
plt.xlabel('Year')
plt.ylabel('Oil consumption per capita (Bl/year)')
plt.title('Oil consumption, Iran vs the World')
plt.legend()
# plt.grid()
```

## Histogram

```python
import numpy as np

plt.figure(figsize=(6,6))
plt.subplot(2, 1, 1)
plt.hist(electricity.loc[:, 2014], bins=np.arange(0, 25000, 1000), label='2014')
plt.ylim((0, 12))
plt.ylabel('Count')
plt.legend()

plt.subplot(2, 1, 2)
plt.hist(electricity.loc[:, 1994], bins=np.arange(0, 25000, 1000), label='1994')
plt.ylim((0, 12))
plt.ylabel('Count')
plt.xlabel('KWh/year/person')
plt.legend()
```


## Comparing Oil and Electricity Growth

```python
plt.figure(figsize=(10, 10))
for country in electricity.index:
    plt.plot(electricity.loc[country, :], oil.loc[country, :], '.', color='gray' ,label=country, alpha=0.2)

plt.plot(electricity.loc['Iran', :], oil.loc['Iran', :], 'r.' ,label=country)


plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.xlabel('Annual electricity consumption (KWh) per capita')
plt.xlabel('Annual oil consumption (BBl) per capita')
plt.title('Oil consumption vs Electricity consumption across countries and Iran')

plt.savefig('oil_vs_electricity.pdf')
plt.savefig('oil_vs_electricity.png', dpi=300)
# plt.legend()
```
