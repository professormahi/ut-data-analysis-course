## Imports

```python
import pandas as pd
import geopandas as gpd
from matplotlib import pyplot as plt
import geoplot

%matplotlib inline
```

## Geomerty Data

```python
gdf = gpd.read_file('../data/geo/irn_admbnda_adm2_unhcr_20190514.shp')
gdf = gdf[['ADM1_FA', 'ADM1_EN','ADM2_FA', 'ADM2_EN', 'geometry']]
```

```python
geoplot.polyplot(gdf)
```

## Read Population Data

```python
df = pd.read_excel('../data/geo/population-urban-95.xlsx', index_col='فهرست ', engine='openpyxl')
df = df.dropna()
```

```python
df
```

```python
df2 = df.groupby('شهرستان').sum()
df2['female_ratio'] = df2['زن'] / df2['مرد']
df2
```

## Data Cleaning

```python
if gdf.loc[0, 'ADM2_FA'].startswith('شهرستان'):
    gdf['ADM2_FA'] = gdf['ADM2_FA'].apply(lambda x: x[8:])
gdf
```

## Join

```python
def regularize_unicode(s):
    s=s.str.replace('ك', 'ک')
    s=s.str.replace('ي', 'ی')
    s=s.str.replace(' ', '')
    s=s.str.replace(chr(8204), '')
    return s

df2.index = regularize_unicode(df2.index)
gdf.ADM2_FA = regularize_unicode(gdf.ADM2_FA)

gdf2=gdf.join(df2.female_ratio, on='ADM2_FA')
gdf2
```

```python
geoplot.choropleth(gdf2,hue='female_ratio', cmap='Spectral', figsize=(14, 8), legend=True, edgecolor='white', linewidth=0.3)
```
