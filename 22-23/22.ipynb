# Joining Tables

```python
df1 = pd.DataFrame({
    'A': ['A11', 'A12'],
    'B': ['B11', 'B12'],
    'C': ['C11', 'C12'],
    'D': ['D11', 'D12'],
}, index=[0, 1])

df2 = pd.DataFrame({
    'A': ['A1', 'A2', 'A3'],
    'B': ['B1', 'B2', 'B3'],
    'C': ['C1', 'C2', 'C3'],
    'D': ['D1', 'D2', 'D3'],
}, index=[1, 2, 3])

display(df1, df2)
```

```python
pd.concat([df1, df2], axis=1, join='inner')
```

# Merge

```python
df1 = pd.DataFrame({
    'id': [81010, 81011, 81012, 81013],
    'DA': [20, 25, 18.75, 10],
    'SI': [16, 17, 20, 13],
    'IL': [17, 13, 14, 11]
})

df2 = pd.DataFrame({
    'id': [81011, 81012, 81013, 81014, 81015],
    'GPA': [16.5, 18, 15, 19.3, 17.7],
    'uni': ['A0', 'A1', 'A2', 'A3', 'A5']
})

display(df1, df2)
```

```python
pd.merge(df1, df2, on='id', how='outer')
```

# Join

```python
df1.join(df2, lsuffix='1', rsuffix='2')
```

```python
df1.set_index('id').join(df2.set_index('id'))
```
