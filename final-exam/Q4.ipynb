```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
df = pd.read_excel('q4-data.xlsx', engine='openpyxl')
display(df)
df = df.drop(columns=['day', 'month', 'year', 'popData2019', 'Cumulative_number_for_14_days_of_COVID-19_cases_per_100000'])
```

```python
plt.figure(figsize=(10, 10))
_ = df.groupby(['continentExp', 'countriesAndTerritories']).sum().reset_index()


for c in _['continentExp'].unique():
    plt.scatter(_[_['continentExp'] == c]['cases'], _[_['continentExp'] == c]['deaths'], alpha=0.2, label=c)
    
plt.legend()
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('Death to Cases Ratio for Covid');
```
